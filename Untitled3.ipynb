```python
import numpy as np
import pandas as pd
import datetime
```

```python
# url for airports csv file
airports_csv = 'https://raw.githubusercontent.com/hadley/nycflights13/master/data-raw/airports.csv'

# url for weather csv file
weather_csv = 'https://raw.githubusercontent.com/hadley/nycflights13/master/data-raw/weather.csv'
```

```python
ndf = pd.read_csv(airports_csv)
ndf.nlargest(5,'lat')
```

Answer 1:
the Northern Most airport in the US is Wiley Post Will Rogers Mem Airport.

```python
edf = pd.read_csv(airports_csv)
edf.nlargest(5,'lon')
```

Answer 2:
Eastport Municipal Airport is the easternmost airport in the United States

```python
wdf = pd.read_csv(weather_csv)
nwdf = wdf.loc[(wdf['year'] == 2013) & (wdf['month'] == 2) & (wdf['day'] == 12)]
nwdf.nlargest(5,'wind_speed')
```
